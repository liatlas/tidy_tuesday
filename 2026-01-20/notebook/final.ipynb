```python
import pandas as pd
pd.set_option("display.max_colwidth", None)
```

```python
data_path = "../data/apod.csv"
df = pd.read_csv(data_path)
df.columns
```

### What Types are Most Common in the Archive?

```python
object_types = df.groupby("media_type").size().reset_index(name='count')
object_types
```

### Are Any Images Posted More Than Once?

```python
image_counts = df[df['media_type'] == 'image'].groupby('url').size().reset_index(name='count').sort_values('count', ascending=False)
dup_images = image_counts[image_counts['count'] > 1]
```

```python
print(f"There are {dup_images.shape[0]} images posted more than once")
```
