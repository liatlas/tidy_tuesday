## How does company size relate to capital size (and how skewed is it)?  

```python
df.head()
```

```python
plt.figure(figsize=(6,6))
plt.title("company size vs. capital stock")

sns.boxplot(df, x="company_size", y="capital_stock")
```

```python
df_no_outliers = df[df['capital_stock'].between(lower_bound, upper_bound)]
df_no_outliers
```

```python
plt.figure(figsize=(6,6))
plt.title('company size vs. capital stock')

sns.violinplot(df_no_outliers, x="capital_stock", y="company_size")
```

```python
print(f"num other: {df_no_outliers[df_no_outliers['company_size'] == 'other'].shape[0]}")
print(f"num micro-enterprise: {df_no_outliers[df_no_outliers['company_size'] == 'micro-enterprise'].shape[0]}")
print(f"num small-enterprise: {df_no_outliers[df_no_outliers['company_size'] == 'small-enterprise'].shape[0]}")
```

```python
company_sizes = ['other', 'micro-enterprise', 'small-enterprise']

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for ax, size in zip(axes, company_sizes):
    sns.histplot(df_no_outliers[df_no_outliers['company_size'] == size], 
                 x='capital_stock', bins=20, ax=ax)
    ax.set_title(f'{size} vs. Capital Stock')

plt.tight_layout() 
plt.show()
```

### All company sizes are very left-skewed 
### Yet company size other, despite having fewer companies, seems to have less skew than the other company sizes
### For non-outliers, company size has little impact on capital stock, although being company size other may be beneficial
